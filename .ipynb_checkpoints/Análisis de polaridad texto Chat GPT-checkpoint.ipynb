# Chat GPT ¿Qué dice sobre el tema?

Para cerrar y haciendo uso de una de las herramientas más disruptivas que ha aparecido durante el bootcamp, vamos a analizar la polaridad de un texto de Chat GPT pidendole a esta Inteligencia Artificial que redacte un escrito sobre el Reskilling Tecnológico Femenino (female technological reskilling), usando una técnica de procesamiento de lenguaje natural determinaremos el sentimiento que aplica esta IA a la hora de otorgar información sobre el tema que nos ocupa. 

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
texto_chat = """Technological reskilling refers to the process of learning new skills or upgrading existing ones to keep up with changes in technology and the job market. Female technological reskilling, therefore, refers to the specific efforts made to help women acquire the necessary skills to succeed in the rapidly changing tech industry.

There is a growing need for women to reskill in technology due to the significant gender gap that exists in the field. Despite some progress in recent years, women remain underrepresented in technology-related fields, particularly in leadership roles. This underrepresentation is largely due to historical and cultural factors that have led to a lack of access and opportunities for women in tech.

To address this issue, there are many initiatives, programs, and resources available to help women reskill in technology. These include:

Coding bootcamps: Many coding bootcamps offer programs specifically tailored to women to provide a supportive environment and encourage more women to enter the tech industry.

Mentorship programs: Mentorship programs connect women with experienced professionals who can offer guidance and support as they develop their skills.

Online courses: There are many online courses available that allow women to learn new skills and technologies at their own pace.

Women in tech organizations: There are many organizations dedicated to supporting women in tech, such as Women Who Code and Girls Who Code, that offer networking events, workshops, and other resources to help women reskill in technology.

Diversity and inclusion initiatives: Many tech companies are implementing diversity and inclusion initiatives to encourage more women to join their teams and provide them with the necessary training and support to succeed.

In summary, female technological reskilling refers to the specific efforts made to help women acquire the necessary skills to succeed in the tech industry. There are many resources available to support women in reskilling, including coding bootcamps, mentorship programs, online courses, women in tech organizations, and diversity and inclusion initiatives."""
```

```python
analisis = SentimentIntensityAnalyzer()
analisis.polarity_scores(texto_chat)
```
