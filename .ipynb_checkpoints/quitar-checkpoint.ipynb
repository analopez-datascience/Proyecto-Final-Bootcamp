```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from bs4 import BeautifulSoup as bs
import warnings
warnings.filterwarnings('ignore')
import requests

import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

# Web Scraping

# Web 1

```python
web_1 = "https://www.cursosfemxa.es/blog/acceso-mujer-trabajo-tecnologico"
r_1 = requests.get(web_1)
r_1.status_code
```

```python
soup_1 = bs(r_1.content)
```

```python
quote_elements_h1 = soup_1.find_all("h1")
quote_elements_h1
```

```python
quote_elements_intro = soup_1.find_all("div", class_="itemIntroText")
quote_elements_intro
```

```python
# quote_elements_body = soup_1.find_all("div", class_="itemFullText")
# quote_elements_body
```

```python
quote_elements_title = soup_1.find_all("h1")
quote_elements_intro = soup_1.find_all("div", class_="itemIntroText")
quote_elements_body = soup_1.find_all("div", class_="itemFullText")

text_list = [quote_elements_title[0].text]

for i in quote_elements_intro:
    text_list.append(i.text)
    
for i in quote_elements_body:
    text_list.append(i.text)

full_text_1 = '\n'.join(text_list)
#print(full_text_1)
```

```python
tokenized_texto_1 = sent_tokenize(full_text_1)
#print(tokenized_texto_1)
```

```python
tokenized_word_1 = word_tokenize(full_text_1)
#print(tokenized_word_1)
```

```python
freqdist_1 = FreqDist(tokenized_word_1)
print(freqdist_1)
```

```python
analisis_1 = SentimentIntensityAnalyzer()
analisis.polarity_scores(full_text_1)
```

# Procesamiento de lenguaje natural

Para cerrar y haciendo uso de una de las herramientas más disruptivas que ha aparecido ultimamente, vamos a analizar la polaridad de un texto del Chat GPT pidendole que redacte un escrito sobre el Reskilling Tecnológico Femenino, con está técnica de procesamiento de lenguaje natural determinaremos el sentimiento que aplica esta IA a la hora de otorgar información sobre el tema que nos ocupa. 

```python
texto_chat = """En la última década, el reskilling tecnológico femenino en España ha experimentado un importante avance. Cada vez son más las mujeres que deciden formarse en competencias digitales y tecnológicas para mejorar sus posibilidades laborales y adaptarse a un mercado laboral en constante evolución.

Este avance se ha visto favorecido por distintas iniciativas impulsadas tanto por empresas como por organismos públicos, que han puesto en marcha programas de formación y capacitación en tecnología y transformación digital dirigidos específicamente a mujeres. A través de estas iniciativas, se busca fomentar la diversidad y la igualdad de oportunidades en el ámbito tecnológico, donde la presencia femenina aún es escasa en comparación con la masculina.

Además, también se han creado espacios de networking y comunidades de mujeres tecnólogas para favorecer el intercambio de conocimientos y experiencias, así como para visibilizar el papel de la mujer en el ámbito tecnológico.

A pesar de los avances, aún existen barreras que dificultan el acceso de las mujeres al ámbito tecnológico, como la falta de referentes femeninos, la brecha de género en la educación en STEM (ciencia, tecnología, ingeniería y matemáticas) o la discriminación laboral. Es por ello que es importante seguir impulsando políticas y medidas que favorezcan la inclusión y la igualdad de oportunidades en el sector tecnológico.

En definitiva, el reskilling tecnológico femenino en la última década en España ha avanzado significativamente, pero aún queda camino por recorrer. La formación y capacitación en competencias digitales y tecnológicas deben seguir siendo una prioridad para fomentar la diversidad y la igualdad de oportunidades en un sector clave para el desarrollo económico y social del país."""
```

```python
analisis = SentimentIntensityAnalyzer()
analisis.polarity_scores(texto_chat)
```

El texto que arroja Chat GTP al preguntarle por el Reskilling Tecnológico Femenino en la última dñecada, destaca por su nuetralidad y sobre todo ausencia de negatividad.
